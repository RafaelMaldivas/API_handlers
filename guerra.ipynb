## Regressão Linear

```python
x = df.drop(['soldados', 'drone', 'helicoptero', 'aeronave', 'interceptador_misseis'], axis=1)
y = df['soldados']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_treino, X_teste, Y_treino, Y_teste = train_test_split(x, y, random_state=101, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression()
```

```python
LR.fit(X_treino, Y_treino)
```

```python
X_treino
```

```python
Y_treino
```

```python
predicao = LR.predict(X_teste)
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(Y_teste,predicao)
```

```python
metrics.mean_squared_error(Y_teste, predicao)
```

# Regressão Linear com 'Prophet'

```python
from fbprophet import Prophet
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 20, 10
```

```python
df_personnel
```

```python
dataset = df_personnel.drop(['day','personnel*', 'POW'], axis=1)
dataset.columns = ['ds', 'y']
```

```python
dataset
```

```python
dataset.plot(title='Dias X Qtd Mortes')
plt.xlabel('Dias', size=26)
plt.ylabel('Mortes', size=26)
plt.legend(title='soldados')
plt.show()
```

```python
dataset['ds'] = dataset['ds'].astype('datetime64[ms]')
```

```python
modelo = Prophet()
```

```python
modelo.fit(dataset)
```

```python
data_futuro = pd.date_range(start='2022-06-28', end='2022-07-29')
```

```python
data_futuro  
```

```python
df = pd.DataFrame(data_futuro)
```

```python
df.columns = ['ds']
df
```

pedindo ao modelo para fazer uma previsão para datas futuras

```python
previsao = modelo.predict(df) 
```

```python
df_personnel.tail(3)
```

```python
previsao = previsao[['ds','yhat_lower',  'yhat','yhat_upper']]
previsao
```

gráfico mostrando na ponta a variação para a previsão de mortes

```python
modelo.plot(previsao)
plt.show()
```

```python
from fbprophet.plot import plot_components_plotly, plot_plotly
```

```python
plot_plotly(modelo, previsao)
```
